```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (16,16)

import numpy as np
import random

primary = np.array([[0,0], [1, 0], [.5, np.sqrt(3)/2]])
internal = np.array([[.5, 0]])

for idx in range(25000):
    da_vez = random.choice(primary)
    temp = np.array([[(da_vez[0]+internal[-1][0])/2, (da_vez[1]+ internal[-1][1])/2]])
    internal = np.append(internal, temp, axis = 0)
    

plt.scatter(np.append(primary, internal, axis = 0)[:,0], np.append(primary, internal, axis = 0)[0:,1], marker = '2')
```
